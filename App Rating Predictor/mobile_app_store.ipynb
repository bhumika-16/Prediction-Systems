```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
app=pd.read_csv("C:\\Users\\Admin\\Desktop\\datasets\\project\\AppleStore.csv")
print(app)
```

```python
x=app.iloc[:,[3,5,6,10,11,12,13,14,15,16]]
y=app.iloc[:,8].values
x
```

```python
app.isnull().sum()
```

```python
app_dummy=pd.get_dummies(app['prime_genre'])
app_dummy1=pd.get_dummies(app['cont_rating'])
app_dummy=pd.concat([app,app_dummy,app_dummy1],axis=1)
app_dummy.isnull().sum()
```

```python
x=app_dummy.iloc[:,3:]
y=app.iloc[:,8].values
x=x.drop(['currency','rating_count_ver','user_rating','user_rating_ver','prime_genre','cont_rating','ver'],axis=1)
x
```

```python
from sklearn import preprocessing
standard=preprocessing.StandardScaler().fit(x)
standard.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
regression_tree = DecisionTreeRegressor(min_samples_split=.05)
regression_tree.fit(x,y)
y_pred=regression_tree.predict(x)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, random_state=45)
regressor.fit(x,y)
y_pred_rf=regressor.predict(x)
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_pred_rf)
```
